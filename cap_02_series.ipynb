```python
import pandas as pd
import numpy as np
```

```python
notas = pd.Series([7.6, 5.0, 8.5, 9.5, 6.4]) #Criada a serie notas por padrão o pandas adicionar o índice de 0 a N-1, onde n é o tamanho da lista e como dado adiciona a lista de notas
print(notas)
```

- Para testar se um rótulo existe em uma Series (mais precisamente, se faz parte de seu vetor de rótulos), a coisa é bem simples: basta utilizar o operador 'in' do Python padrão (que retornará True ou False ).

- Se o que você deseja é verificar se um ou mais valores estão em uma Series (mais precisamente, se fazem parte de seu vetor de dados), precisará utilizar o método 'isin()' da pandas.

- Este método recebe como entrada uma lista contendo um ou mais valores. Como saída, é gerada uma Series com dtype bool que terá o valor True para todos os rótulos associados a valores da lista.

- Entretanto, em grande parte das situações práticas, não precisaremos fazer uso desse recurso. 
- Isso porque, em geral, as operações da pandas podem ser executadas através do mecanismo conhecido como computação vetorizada (vectorization). 
- Neste processo, as operações são realizadas sobre cada elemento da Series automaticamente, sem a necessidade de programar um laço com o comando 'for'. 

```python
s1 = pd.Series([2,4,6])
s2 = pd.Series([1,3,5])
print(s1)
print('-' * 20)
print(s2)
```

```python
soma = s1 + s2 #Com o pandas podemos realizar a soma da serie s1 mais a serie s2 gerando a serie soma
print(soma)
```

```python
vezes = s1 * s2 #Com o pandas podemos realizar a multiplicação da serie s1 vezes a serie s2 gerando a serie vezes
print(vezes)
```

```python
raiz_quad = np.sqrt(s1) #Utilizando a biblioteca Numpy geramos a serie raiz_quad com os valores da raiz quadrada da serie s1
print(raiz_quad)
```

- três operações são efetuadas:
    - Soma de s1 e s2 , obtida por s1 + s2 . Veja que a soma é feita entre pares de elementos de s1 e s2 que ocupam a mesma posição (compartilham o mesmo índice);
    - Multiplicação de todos os elementos de s1 por s2, obtida por s1 * s2 ;
    - Raiz quadrada dos elementos de s1, obtida por np.sqrt(s1).

```python
verde = pd.Series({'BR':1, 'FR': 0, 'IT':1, 'UK': 0})
azul = pd.Series({'AR':1, 'BR':1, 'FR': 1, 'IT':0, 'UK': 1})
```

```python
soma = verde + azul
print(soma)
```

- como 'AR' existe em azul , mas não tem um correspondente em verde, a pandas interpreta que o valor de 'AR' em verde é desconhecido. 
- E o resultado de qualquer operação aritmética envolvendo um número conhecido e um desconhecido resultará sempre em desconhecido (ou seja, em NaN, conceito utilizado pela pandas para marcar valores nulos/ausentes (missing) ou desconhecidos).
- Ou seja, 1 + desconhecido = desconhecido.

```python
desconhecido = pd.isnull(soma) #  O método isnull() recebe como entrada uma serie, gerando como saída uma outra serie com dtype bool , que indica quais dos rótulos estão associados a valores nulos ( NaN )
print(desconhecido)
```

```python
dias = ['10/02/2019', '11/02/2019','12/02/2019','13/02/2019',
'14/02/2019','15/02/2019']
temp_max = [31,35,34,28,27,27]

serie_temporal = pd.Series(data = temp_max, index = dias)
print(serie_temporal)
```

```python
serie_temporal.index = pd.to_datetime(serie_temporal.index, format='%d/%m/%Y') #converte o tipo do índice para datetime 
print(serie_temporal)
```

O truque para transformar os valores deste índice para datetime consistiu no emprego do método to_datetime() com dois parâmetros:
- O nome do vetor de índices a ser convertido (em nosso caso, serie_temporal.index).
- O format='%d/%m/%Y': o parâmetro format é utilizado para que possamos informar a pandas o formato da data (date string). Para indicar este formato, foi preciso utilizar as máscaras %d (dia), %m (mês) e %Y (ano com quatro dígitos) combinadas com o caractere /. No mundo Python, essas máscaras são chamadas de diretivas (directives). 

```python
moedas = ['Peso', 'Real', 'Euro', 'Euro', 'Libra']
paises = [['América','América','Europa','Europa','Europa'],['AR','BR','FR','IT','UK']]

paises = pd.Series(data = moedas, index = paises)
print(paises)
```

```python
print(paises['América']) #indexação pelo nível 1 (nome do continente). O resultado é uma Series contendo as siglas e moedas dos países da América.
```

```python
print(paises[:,'IT']) #indexação pelo nível 2 (sigla do país). Como resultado, é retornada uma Series com um único elemento, em que o rótulo é o nome do continente e valor é o nome da moeda da Itália.
```

```python
print(paises['Europa', 'IT']) #indexação por ambos os níveis. Como resultado, será retornado 'Euro' (valor string), pois este é o valor do elemento cujo nível 1 do índice é 'Europa' e o nível 2 'IT'.
```

- A Series paises foi criada com um vetor de rótulos indexado em dois níveis. 
    - No primeiro nível temos o nome do continente ('América' ou 'Europa')
    - No segundo nível, a sigla do país ('AR', 'BR', 'FR', 'IT' e 'UK'). 
    - Com isso, torna-se possível indexar os dados de três formas: apenas pelo nível 1, apenas pelo nível 2 ou por ambos.
